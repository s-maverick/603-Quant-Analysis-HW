## **Hypothesis Testing**  
  
Hypothesis testing is strongly tied to the concept of confidence intervals.
For example, if a hypothesized population mean $\mu$ is not within the 95% confidence interval, then we can reject the null hypothesis that the population mean equals $\mu$ at the 5% significance level.

Elements of a Statistical Test:  
1. Null Hypothesis, $H_{0}$  
2. Alternative Hypothesis, $H_{\alpha}$  
3. Test Statistics  
4. Rejection Region
5. Conclusion  


### **`Null Hypothesis`**
  
The null hypothesis, denoted $H_{0}$, is a statement to be tested. The null hypothesis is usually a statement of no change, no effect or no difference and is assumed true until evidence indicates otherwise. The parameter value of $H_{0}$ is a single point (because of the equality sign).
  
### **`Alternative Hypothesis`**
  
The alternative hypothesis, denoted by $H_{\alpha}$ or $H_{1}$, is a statement that we are trying to find evidence (from the sample data) to support. The parameter values of $H_{\alpha}$ form a region, i.e. there are multiple values (because of inequality signs). Alternative Hypothesis can either be one-sided {i.e. either left-tailed or right tailed} or two-sided.
  
### **`Test Statistics`**
  
The Test statistic is a function of the sample measurements on which the statistical decision will be based. It usually contains data such as significance level, mean, median, standard deviation, standard error, type of distribution used to make calculations.
  
### **`Rejection Region`**
  
The Rejection Region `(RR)`, specifies the values of the test statistic for which the null hypothesis will be rejected in favor of the alternate hypothesis. In other words, if the test statistic falls in the `RR`, then we reject the null hypothesis. We also determine the $p-value$ in Rejection Region. `p-value` is the probability that test statistic equals the observed data or data even more extreme as predicted by $H_{\alpha}$. It is calculated by assuming $H_{0}$ to be true.  
  
### **`Conclusion`**  
  
After conducting a series of tests and calculating values from distributions, p-value, and test statistics, we should be able to reject or fail to reject the null hypothesis based on feedback and insights from your sample data.  
  
If our `p-value` is less than the pre-determined significance ($\alpha$) level, typically **5%**, we reject null hypothesis and conclude that we do have statistical evidence for the alternative hypothesis is true. If our `p-value` is greater than the significance level, we fail to reject null hypothesis. Hence, we do not have enough statistical evidence to draw conclusions.

### **`One Sample t-test`**  
  
The `t.test()` function has another additional argument: `mu`. `mu` determines a number indicating the true value of the mean (or a difference in means if it is a two sampled test). The default value is 0 but you can specify whatever you want based on the needs of your application.    
  
Let's test whether the population mean Height of Cathedral is 85 ($H_0: \mu = 85$) against the alternative hypothesis that it is not ($H_a: \mu \neq 85$).  

```r
t.test(cathedral$Height, mu = 85)
```

After executing this code, we can see that the `t.test()` function displays information about the alternative hypothesis. For the above code, we can see that we have a very small p-value, which indicates that we can reject the null hypothesis at most conventional levels, such as 5% or 1%, in favor of the alternative that the true mean is not 85. The 95% confidence interval is automatically calculated. The fact that it does not contain 85 shows the same thing another way. In the `t.test()` function, we can change the alternative hypothesis to test on the specified data by adding another argument: `alternative`. `alternative` by default is `two.sided` but it's values can either be `greater`, `less`, or `two.sided`.  
  
Let's test whether the true mean of the Height population of Cathedral is less than 85.

```r
t.test(cathedral$Height, mu = 85, alternative = 'less')
```

Note that, when we use 'less', the p-value is half of what it was when we used 'two.sided.' This is because, in a one sided test, only extreme values only on one side of the null distribution are considered evidence against the null. Intuitively, we have a sample with a mean of 75.16 and we are considering whether the population mean is 85. We are more confident when concluding that the population must be lower than 85 (a smaller p-value) than when concluding that the population mean is not 85 (a still small, but somewhat larger p-value than the former). For the latter, deviations in either direction could be evidence.

### `prop.test()`  
  
`prop.test()` is the **Z-Test of Proportions with Continuity Correction**. It is similar to the t-test, but is suited for tests about proportions. `prop.test()` has 3 main arguments:`x`, `n`, and `p`.
There can be additional requirements such as:`alternative`, `conf.level`, and `correct`.  
  
Here,  
1. `x`: vector of counts of successes
2. `n`: vector of counts of trials  
3. `p`: vector of probabilities of success  
4. `alternative`: either `two.sided`, `greater`, or `less`
5. `conf.level`: denotes the confidence level chosen  
6. `correct`: a logical value denoting where to apply continuity correction

Let's do a simulated example:

```r
set.seed(0)
heads <- rbinom(1, size = 100, p = 0.3)
prop.test(heads, 100)
```

Even though we simulated with a probability of heads equal to 0.3 (30%), the 95% percent confidence interval for the prop.test() does include 0.5. We are unable to reject the null hypothesis that the true probability is 0.5 (which is the default null hypothesis in a one sample test in `prop.test()`) at the 5% significance level.
