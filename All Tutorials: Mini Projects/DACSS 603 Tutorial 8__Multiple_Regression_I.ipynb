## Prediction and Inference

Linear regression models are typically used for two main purposes: prediction and inference.
Prediction is the main concern in the machine learning tradition: construct a model and apply it to new cases to get a predicted value for the outcome variable.
Social sciences are more concerned with explanation of mechanisms, causes and effects, and statistical significance and inference---usually through null hypothesis significance testing on one or more explanatory variables.

### Prediction with predict()

R has various functions that can be used for these purposes, and we'll just take a look at two here.

The `predict()` function can be used to make predictions using a linear regression model for new cases.

Fit the very first model again (with female life expectancy as the outcome variable and with GDP per capita and share of urban population as explanatory variables), but assign the output to an object called `fit`. (Assign the output from `lm()` to `fit`, **not** the output from `summary(lm(...))`.)

```r
# Write your code here
fit = lm(lifeExpF ~ ppgdp + pctUrban, data = UN)
```

If you simply use the `fit` object as the only input to the `predict()` function, it will produce the predicted female life expectancy for the observations in the model (which are also used to calculate residuals).
Let's try it out:

```r
head(predict(fit), n = 10)
```

Importantly, `predict()` can be used to make predictions about new data points that we supply.
Let's create one that we can then use:


```r
set.seed(3)

df <- data.frame(ppgdp = sample(UN$ppgdp, size = 10, replace = T),
           pctUrban = sample(UN$pctUrban, size = 10, replace = T))

df
```

We used `sample()` to sample from the existing values.
This will ensure that we have plausible input values that are not out of the range that was used when fitting the model.
But we will have many combinations of the input variables that the model did not use.

Now, use `df`, as the second argument to the `predict()` function to get predictions (while still using `fit` as the first argument.)

```r
# Write your code here
predictions <- predict(fit, newdata = df)
head(predictions, n = 10)
```


### Statistical Inference

The main tool of statistical inference for a linear model is something we have already seen: the `summary()` function.
It reports the standard error, t value, p value for each coefficient and even marks each coefficient for different levels of significance.
It also reports an F-statistic and a p-value associated with that, which can be used to assess the explanatory power of the predictors altogether.

Let's revisit `summary()`. Call the `summary()` function on the `fit` object.


```r
# Write your code here
summary(fit)
```

The results show p-values that are very small, providing strong evidence against the null hypothesis that the real "population" partial coefficients for `ppgdp` and `pctUrban` are zero.
We have reason to believe that these coefficients are not zero (i.e. to believe in the alternative hypothesis).
The three stars (***) indicate statistical significance at the 0.01 level.

However, as we have previously seen, looking at confidence intervals reveals much more information than the hypothesis test alone.
For example, it would show us what the range of plausible values for these variables are.
Right now, we believe they are not zero, but we are not sure about how close to or far from zero they can plausibly be.

We can use the `confint()` function to construct the relevant confidence intervals.
Let's construct a confidence interval for the GDP per capita variable's coefficient.

```r
confint(fit, 'ppgdp')
```

The `confint()` function constructs a 95% confidence interval by default. But that can be changed:

```r
confint(fit, 'ppgdp', level = 0.99)
```
